#### **Phenomenon**

- This analysis examines non-biogenic carbon dioxide (CO₂), the portion of greenhouse gases produced by non-renewable carbon sources such as coal, oil, natural gas, and petroleum products. These emissions are human-caused and primarily linked to industrial processes. In 2022, industrial activity accounted for about 23% of total U.S. greenhouse gas emissions, making this a major contributor to national emission totals.
- The project focuses on year-by-year non-biogenic CO₂ emissions across U.S. states and territories, with totals normalized by land area to account for differences in geographic size and to measure emission intensity. This distinction allows for fair comparison between large states with high total emissions and smaller, more industrialized states with concentrated emissions.

#### **Goal**
- The goal of this project is to quantify and compare non-biogenic CO₂ emissions across U.S. states and territories over time, revealing how total emissions and emission intensity vary by geography and industrial activity. To achieve this, the analysis applies several statistical techniques: the Empirical Cumulative Distribution Function (ECDF) to visualize cumulative emission levels across states, Kernel Density Estimation (KDE) to identify where emissions are most concentrated, and bootstrapping to evaluate the stability and reliability of these observed patterns through repeated resampling.

#### **Main Features of the Analysis**
- Results show that larger states tend to have higher total emissions due to scale, while smaller, densely industrialized states demonstrate higher emissions per square mile. ECDF and KDE analyses reveal right-skewed distributions, with most states producing relatively low emissions and a few contributing disproportionately high levels. Bootstrapping confirms these distribution shapes are stable, providing confidence in the robustness of the observed trends.
