#### **Phenomenon**

- This project models greenhouse gas (GHG) emissions from stationary combustion and process units in U.S. industrial facilities. It examines how differnt technologies and fuels influence emissions of carbon dioxide (CO2), methan (CH4), and nitrous oxide (N2O), three gases with varying global warming potentials that significantly contribute to climate change.

#### **Goal**
- The goal is to quantify and compare emissions across unit types, fuels, industries, and locations to identify which configurations are most carbon-intensive.By understanding these relationships, the model provides insights that can guide emission-reduction strategies and sustainable industrial planning.

#### **Dataset**
- The dataset includes facility and unit-level data from the EPA's Greehouse Gas Reporting Program. Key fields capture unit type, fuel classification, maximum heat input capacity (mmBTU/hr), industry sector (NAICS code), state and reporting year. Emissions are reported for non-biogenic CO2, biogenic CO2, CH4, and N2O, along with corresponding and geographic identifiers.

#### **Key findings from group-bys**
- Electricity generators and “other combustion sources” (OCS) dominate total CO2 emissions.
- Coal-fired units and petroleum-based fuels exhibit the highest CH4 and N2O emissions.
- Texas, Pennsylvania, and Ohio lead in total CO2 emissions due to large energy and industrial sectors.

These findings confirm that unit technology, fuel type, and regional industry mix are the primary drivers of facility-level GHG emissions.
