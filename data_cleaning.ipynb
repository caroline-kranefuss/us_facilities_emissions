```python
import pandas as pd
```

### **unit data sheet**

```python
unitdata_df = pd.read_excel('/Users/sophiekim/Desktop/msds/ds 5030/project/emissions_by_unit_and_fuel_type_c_d_aa.xlsb', engine='pyxlsb', skiprows=6, sheet_name='UNIT_DATA')

unitdata_df.head()
```

```python
unitdata_df.columns
```

```python
unitdata_df['Unit Type'] = unitdata_df['Unit Type'].astype(str).str.strip()
```

```python
# group by industry type & summing by all the emissions

emissions_cols = ['Unit CO2 emissions (non-biogenic) ', 'Unit Methane (CH4) emissions ',
       'Unit Nitrous Oxide (N2O) emissions ',
       'Unit Biogenic CO2 emissions (metric tons)']


emissions_by_unit = unitdata_df.groupby('Unit Type')[emissions_cols].sum()
emissions_by_unit
```

```python
unitdata_df['Facility Name'].value_counts()
```

```python
emissions_by_facility = unitdata_df.groupby('Facility Name')[emissions_cols].sum()
emissions_by_facility
```

```python
emissions_by_state = unitdata_df.groupby('State')[emissions_cols].sum()
emissions_by_state
```
